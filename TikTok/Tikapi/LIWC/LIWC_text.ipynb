## SImilarity LIWC

```python
#functions for similarity and bipartite graph
def build_pr_matr(matrix_pr, names, filename, cutoff):
    pr_matr = np.zeros((len(names),len(names)))
    for i in range(len(names)):
        for j in range(i+1,len(names)):
            X = matrix_pr[i]
            Y = matrix_pr[j]
            val = stats.pearsonr(X,Y)[0]
            if (val >= cutoff):
                pr_matr[i,j] = val
                
    pr_matr_= pd.DataFrame(pr_matr, columns = names, index = names)
    pr_matr_.to_csv(filename)
    
    return(pr_matr)

def build_sim_matr(matrix_freq, names, filename, cutoff):
    tfidfTran = TfidfTransformer(norm="l2")
    tfidfTran.fit(matrix_freq)

    tfidf_matrix = tfidfTran.transform(matrix_freq)

    cos_similarity_matrix = (tfidf_matrix * tfidf_matrix.T).toarray()

    cos_similarity_matrix[cos_similarity_matrix < cutoff] = 0.
    
    cos_matr = pd.DataFrame(cos_similarity_matrix, columns = names, index = names)
    
    cos_matr.to_csv(filename)
    
    return(cos_similarity_matrix)

def bipartite(matr, names, inf_names, filename):
    
    N = len(inf_names)
    
    matr[:N,:N] = 0.
    
    matr[N:,N:] = 0.
    
    matr_ = pd.DataFrame(matr, columns = names, index = names)
    matr_.to_csv(filename)
```
