```python
import pandas as  pd
from IPython.display import display
```

## We are using LIWC 2015

```python
#read excel file with liwc info
df = pd.read_excel("LIWC/LIWC_emoji_text.xlsx") 
```

```python
display(df)
```

```python
df_comments = df.loc[df['Source (C)'] == "comm"]
display(df_comments)
df_text = df.loc[df['Source (C)'] == "text"]
display(df_text)
```

# Comments part

### Calculating the overall mean of interesting columns

```python
##calculate some interesting values

#calculate mean of positive and negative emotions in comments
feature_list = [
    "WC","Tone","Clout","affect","posemo","negemo","anx","anger","sad", "social", "family","female", "male",
    "body","health", "achieve", "work", "money", "relig", "i", "we"
]
df_means = df_comments[feature_list].mean()
display(df_means)
```

### Calculating the means for each person

```python
df_people_means = df_comments.groupby("Source (B)")[feature_list].mean()
display(df_people_means)
```

### Calculating the means for each polititian


```python
politicians_names = ["Berlusconi","Conte","Calenda","Meloni","Partito_Democratico", "Renzi", "Salvini","Luca_Carabetta", "manueltuziofficial"]
```

```python
df_politicians_means = df_people_means.loc[politicians_names]
display(df_politicians_means)
```

### Calculating the means for each influencer

```python
df_influencers_means = df_people_means.loc[set(df_people_means.index) - set(politicians_names)]
display(df_influencers_means)
```

### Calculating overall means for each features for politians and influencer

```python
overall_means_politicians = df_politicians_means.mean()
print("Policians:")
display(overall_means_politicians)

overall_means_influencers = df_influencers_means.mean()
print("Influencers:")
display(overall_means_influencers)
```

### Plotting some results

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches


fig = plt.figure() # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.4

overall_means_politicians.plot(kind='bar', color='red', ax=ax, width=width, position=1,label="Politicians")
overall_means_influencers.plot(kind='bar', color='blue', ax=ax2, width=width, position=0,label="Influencers")
ax.set_ylabel('mean value')
ax.set_title('mean LIWC values in the comments')


PO = mpatches.Patch(color='blue', label='Politicians')
IN = mpatches.Patch(color='red', label='Influencers')
plt.legend(handles=[PO,IN], loc=1)
plt.show()
```
