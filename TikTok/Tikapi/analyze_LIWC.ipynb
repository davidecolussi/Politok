## We are using LIWC 2015

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as  pd
from IPython.display import display
```

```python
df_comms = pd.read_excel("LIWC/Comm_LIWC_scaled.xlsx") 
df_comms.rename(columns = {'Unnamed: 0':'Name'}, inplace = True)
display(df_comms)

df_text = pd.read_excel("LIWC/Text_LIWC_scaled.xlsx") 
df_text.rename(columns = {'Unnamed: 0':'Name'}, inplace = True)
display(df_text)
```

```python
politicians_names = ["Berlusconi","Conte","Calenda","Meloni","Partito_Democratico", "Partito", "Renzi", "Salvini","Luca_Carabetta", "Luca","manueltuziofficial"]
summary = ["Analytic", "Clout", "Authentic", "Tone"]
pronouns = ["i", "we", "you", "shehe", "they"]
emotions = ["posemo", "negemo", "anx", "anger", "sad"]
```

#### Distinguish politicians and influencers

```python
df_comms_politicians = df_comms.loc[df_comms['Name'].isin(politicians_names)]
df_comms_influencers = df_comms.loc[~df_comms['Name'].isin(politicians_names)]
print("Number of politicians for the comments:",len(df_comms_politicians))
print("Number of influencers for the comments:",len(df_comms_influencers))

df_text_politicians = df_text.loc[df_text['Name'].isin(politicians_names)]
df_text_influencers = df_text.loc[~df_text['Name'].isin(politicians_names)]
print("Number of politicians for the text:",len(df_text_politicians))
print("Number of influencers for the text:",len(df_text_influencers))
```

```python
#SUMMARY
df_comms_politicians_summary = df_comms_politicians[summary]
#display(df_comms_politicians_summary)
means_politicians_summary = df_comms_politicians_summary.mean()
#display(means_politicians_summary)

df_comms_influencers_summary = df_comms_influencers[summary]
#display(df_comms_influencers_summary)
means_influencers_summary = df_comms_influencers_summary.mean()
#display(means_influencers_summary)


df_text_politicians_summary = df_text_politicians[summary]
#display(df_text_politicians_summary)
means_text_politicians_summary = df_text_politicians_summary.mean()
#display(means_text_politicians_summary)

df_text_influencers_summary = df_text_influencers[summary]
#display(df_text_influencers_summary)
means_text_influencers_summary = df_text_influencers_summary.mean()
#display(means_text_influencers_summary)
```

```python
#SUMMARY PLOT COMMENTS
fig = plt.figure() # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.2

means_politicians_summary.plot(kind='bar', color='red', ax=ax, width=width, position=1,label="Politicians")
means_influencers_summary.plot(kind='bar', color='blue', ax=ax2, width=width, position=0,label="Influencers")
ax.set_ylabel('mean value')
ax.set_title('mean LIWC values in the comments')

PO = mpatches.Patch(color='blue', label='Politicians')
IN = mpatches.Patch(color='red', label='Influencers')
plt.legend(handles=[PO,IN], loc=1)
plt.show()


#SUMMARY PLOT TEXT
fig = plt.figure() # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.2

means_text_politicians_summary.plot(kind='bar', color='red', ax=ax, width=width, position=1,label="Politicians")
means_text_influencers_summary.plot(kind='bar', color='blue', ax=ax2, width=width, position=0,label="Influencers")
ax.set_ylabel('mean value')
ax.set_title('mean LIWC values in the video text')


PO = mpatches.Patch(color='blue', label='Politicians')
IN = mpatches.Patch(color='red', label='Influencers')
plt.legend(handles=[PO,IN], loc=1)
plt.show()
```
